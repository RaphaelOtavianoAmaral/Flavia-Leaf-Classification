# Pré-processamento e Extração de características -  Dataset Flavia Leaf

```python
from modules.preprocess import *
from modules.feature_extraction import *
from modules.dataset import *
from pandas import DataFrame
from skimage.io import imread
```

Localizando a pasta Leaves e listando as imagens encontradas

```python
dataset_folder = load_dataset_dir(r"Leaves")
img_files = load_img_files_names(dataset_folder)
```

Criando função para pré-processar e extrair as características de cada imagem encontrada

```python
def generate_dataset(images_path: list[Path] | None):
    if images_path is None: return None

    features_names = [
        "Área","Perímetro","Eixo Maior","Eixo Menor",
        "Excentricidade","Solidez","Circularidade",
        "Compacidade","Alongamento","Redondeza","Num Cantos"
    ]

    dataset_features_vector = []
    dataset_indexes = []
    
    for image_path in images_path:
        name = image_path.name.split("/")[0]
        try:
            image = imread(image_path)
            preprocessed = preprocess_image(image)
            features_vector = extract_features(preprocessed)
            if len(features_vector) == len(features_names):
                dataset_features_vector.append(features_vector)
                dataset_indexes.append(name)
            else:
                print(f"Imagem {name} mal processada. {len(features_vector)} != {len(features_names)}")
        except Exception as e:
            print(f"Erro no pré processamento e extração de características da imagem {name}.\n{e}")
        
    
    dataset_DF = DataFrame(dataset_features_vector,columns=features_names,index=dataset_indexes)

    print(dataset_DF.head())

    dataset_DF.to_csv("dataset_features.csv")
        
```

```python
generate_dataset(img_files)
```
